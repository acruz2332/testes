```python
import pandas as pd
```

```python
df_test = pd.read_csv('healthkathon_afterpreprocessing.csv')
df_test = df_test.sort_values(by='id', ascending=True)
df_test
```

```python
X_test = df_test.drop(columns=['Unnamed: 0', 'label'])
X_test
```

```python
y_test = df_test['label']
y_test
```

```python
df_train = pd.read_csv('trainset_randomoversampling2.csv')
```

```python
df_train = pd.read_csv('trainset_SMOTE.csv')
```

```python
df_train
```

```python
df_X = df_train.drop(columns=['label', 'Unnamed: 0'])
df_X
```

```python
df_y = df_train['label']
df_y
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(df_X, df_y)
```

```python
ypred_nb = nb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(classification_report(y_test, ypred_nb))
print(confusion_matrix(y_test, ypred_nb))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(df_X, df_y)
```

```python
ypred_knn = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(classification_report(y_test, ypred_knn))
print(confusion_matrix(y_test, ypred_knn))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada=AdaBoostClassifier()
```

```python
ada.fit(df_X, df_y)
```

```python
ypred_ada = ada.predict(X_test)
```

```python
print(classification_report(y_test, ypred_ada))
print(confusion_matrix(y_test, ypred_ada))
```

```python
from xgboost import XGBClassifier
sgb = XGBClassifier()
```

```python
sgb.fit(df_X, df_y)
```

```python
ypred_sgb = sgb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(classification_report(y_test, ypred_sgb))
print(confusion_matrix(y_test, ypred_sgb))
```
