```python
import pandas as pd
data1 = pd.read_csv('/content/gdrive/MyDrive/healthkathon/sampling_healtkathon2022/sampling_healtkathon2022.csv')
data1
```

```python
data1 = data1.sort_values(by='id', ascending=True)
data1
```

```python
data1['label'].value_counts()
```

```python
df = data1
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['jenkel'].mode()
```

```python
df['jenkel'] = df['jenkel'].fillna('P')
```

```python
df['diagfktp'].mode()
```

```python
df['diagfktp'] = df['diagfktp'].fillna('Z09.8')
```

```python
havenull = ['pisat', 'politujuan', 'biaya','jenispulang', 'kdsp', 'kdsa', 'kdsr', 'kdsi', 'kdsd']
for i in havenull:
  df[i] = df[i].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['kdsa'] = df['kdsa'].replace(0, '0')
df['kdsp'] = df['kdsp'].replace(0, '0')
df['kdsr'] = df['kdsr'].replace(0, '0')
df['kdsi'] = df['kdsi'].replace(0, '0')
df['kdsd'] = df['kdsd'].replace(0, '0')
df['politujuan'] = df['politujuan'].replace(0, '0')
```

```python
df['kdsa'] = le.fit_transform(df['kdsa'])
df['kdsp'] = le.fit_transform(df['kdsp'])
df['kdsr'] = le.fit_transform(df['kdsr'])
df['kdsi'] = le.fit_transform(df['kdsi'])
df['kdsd'] = le.fit_transform(df['kdsd'])
df['jenkel'] = le.fit_transform(df['jenkel'])
df['typefaskes'] = le.fit_transform(df['typefaskes'])
df['politujuan'] = le.fit_transform(df['politujuan'])
df['diagfktp'] = le.fit_transform(df['diagfktp'])
df['cbg'] = le.fit_transform(df['cbg'])
```

```python
df['tgldatang'] = pd.to_datetime(df['tgldatang'], format = '%Y-%m-%dT%H:%M:%S.%fZ', errors='coerce')
df['tglpulang'] = pd.to_datetime(df['tglpulang'], format = '%Y-%m-%dT%H:%M:%S.%fZ', errors='coerce')
```

```python
df['tgldatang_year'] = df['tgldatang'].dt.year
df['tgldatang_month'] = df['tgldatang'].dt.month
df['tgldatang_day'] = df['tgldatang'].dt.day
df['tgldatang_hour'] = df['tgldatang'].dt.hour
df['tgldatang_minute'] = df['tgldatang'].dt.minute

df['tglpulang_year'] = df['tglpulang'].dt.year
df['tglpulang_month'] = df['tglpulang'].dt.month
df['tglpulang_day'] = df['tglpulang'].dt.day
df['tglpulang_hour'] = df['tglpulang'].dt.hour
df['tglpulang_minute'] = df['tglpulang'].dt.minute
```

```python
df.drop(columns=['tgldatang', 'tglpulang'], axis=1, inplace=True)
```

#oversampling

**SMOTE**


```python
from imblearn.over_sampling import SMOTE
X_resampled2, y_resampled2 = SMOTE().fit_resample(X_train, y_train)
```

```python
y_resampled2.value_counts()
```

```python
X_resampled2['label'] = y_resampled2
```

```python
X_resampled2.to_csv('/content/gdrive/MyDrive/healthkathon/sampling_healtkathon2022/trainset_SMOTE.csv')
```

**random sampling**

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=12)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
```

```python
y_resampled.value_counts()
```

```python
X_resampled['label'] = y_resampled
```
